# 3 处理 RNA.csv 文件

```python
import numpy as np
import pandas as pd
```

```python
# 读取文件
rna_file = pd.read_csv('../RNA.csv', index_col=0)
```

```python
rna_file.head()
```

```python
# 去掉 0 值 和 NA 值 超过70% 的行和列
drop_index = [ ]
drop_columns = [ ]
for i in range(len(rna_file)):
    if((sum(rna_file.iloc[i, :] == 0) + rna_file.iloc[i, :].isna().sum()) >= len(rna_file.iloc[0, :]) * 0.7):
        drop_index.append(rna_file.index[i])

for i in range(len(rna_file.columns)):
    if((sum(rna_file.iloc[:, i] == 0) + rna_file.iloc[:, i].isna().sum()) >= len(rna_file.iloc[:, 0])*0.7):
        drop_columns.append(rna_file.columns[i])

rna_file.drop(drop_index, axis=0, inplace=True)
rna_file.drop(drop_columns, axis=1, inplace=True)
```

```python
rna_file.describe()
```

```python
# 将剩下的 NA 值以该行(所有样本该基因)的均值替代
for i in range(len(rna_file)):
    rna_file.iloc[i, :].fillna(value=rna_file.iloc[i, :].mean(), inplace=True)
```

```python
# 确认是否还存在 NA 值
sum(rna_file.isna().sum())
```

```python
rna_file
```

```python
# 将每个基因的取值缩放到 [-1, 1]
for i in range(len(rna_file)):
    rna_file.iloc[i, :] /= rna_file.iloc[i, :].max()
```

```python
sum(rna_file.isna().sum())
```

```python
# 将区间 [-1, 1] 映射到 [0, 1] ([-1, 1] + 1 ) / 2 = [0, 1]
for i in range(len(rna_file)):
    rna_file.iloc[i, :] = (rna_file.iloc[i, :] + 1) / 2
```

```python
sum(rna_file.isna().sum())
```

```python
rna_file.T.describe()
```

```python
# 保存文件
rna_file.to_csv('RNA.csv')
```
