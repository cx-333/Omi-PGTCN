# 1 **.gz 文件转 **.csv 文件

```python
import numpy as np
import pandas as pd
import csv
import os
import gzip
```

```python
def gzip_to_csv(f_path:str, store_path:str) -> None:
    """
    f_path: 待转换文件目录
    store_path: 存储目录
    return: None
    """
    print("loading the input data...\n beginning to convert gz file to csv file...")
    with gzip.open(f_path, 'rb') as f:
        text = f.readlines()
        c = open(store_path, 'w', encoding='utf-8', newline="")
        csv_writer = csv.writer(c)
        for line in text:
            line = line.decode().split()
            csv_writer.writerow(line)
        c.close()
        print("converted successfully...")
```

```python
dna_path = '../raw_data/jhu-usc.edu_PANCAN_HumanMethylation450.betaValue_whitelisted.tsv.synapse_download_5096262.xena.gz'
rna_path = '../raw_data/PANCAN_IlluminaHiSeq_RNASeqV2.geneExp.xena.gz'
rppa_path = '../raw_data/TCGA-RPPA-pancan-clean.xena.gz'
store_dna = '../DNA.csv'
store_rna = '../RNA.csv'
store_rppa = '../RPPA.csv'

gzip_to_csv(rppa_path, store_rppa)
gzip_to_csv(rna_path, store_rna)
gzip_to_csv(dna_path, store_dna)
```

# 2 处理 RPPA.csv 文件

```python
# 读取文件
rppa_file = pd.read_csv('../RPPA.csv', index_col=0)
```

```python
rppa_file.describe()
```

```python
# 去掉全为 NA 和 超过70% NA 的行 和 列  
rppa_file.dropna(axis='index', how='all', thresh=int(7754*0.7), inplace=True)   # 行
rppa_file.dropna(axis='columns', how='all', thresh=int(223*0.7), inplace=True)  # 列
```

```python
rppa_file.describe()
```

```python
# 将剩下的 NA 值以该行的均值替代
for i in range(len(rppa_file)):
    rppa_file.iloc[i, :].fillna(value=rppa_file.iloc[i, :].mean(), inplace=True)
```

```python
# 确认是否还存在 NA 值
sum(rppa_file.isna().sum())
```

```python
rppa_file.describe()
```

```python
# 将每个基因的取值缩放到 [-1, 1]
for i in range(len(rppa_file)):
    rppa_file.iloc[i, :] /= np.max(rppa_file.iloc[i, :])
```

```python
# 将区间 [-1, 1] 映射到 [0, 1] ([-1, 1] + 1 ) / 2 = [0, 1]
for i in range(len(rppa_file)):
    rppa_file.iloc[i, :] = (rppa_file.iloc[i, :] + 1) / 2
```

```python
# 保存文件
rppa_file.to_csv('RPPA.csv')
```
